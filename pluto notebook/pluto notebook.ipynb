### 1.	Import pandas and open the csv file as df1.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('pluto_18v2_1.csv')
boroughs = {'BX': 2, 'BK': 3, 'MN': 1, 'QN': 4, 'SI': 5}
df1 = df1[df1['borough'].isin(boroughs.keys())]
df1 = df1.dropna(subset=['ct2010'])
df1.head()
```

### 2.	In df1, create a new variable that includes the borough census tract:
- a.	Use a replace function on the Borough variable so that the following numbers replace the text: BX:2, BK:3, MN:1, QN:4, SI: 5.
- b.	Then create a new variable, btract, that concatenates this variable to the Tract 2010 variable. Your resulting variable should be a 7 digit string: e.g. 1001001.


```python
df1['btract'] = df1['borough'].replace(boroughs).astype(str) + df1['ct2010'].astype(str) 
```

### 3.	Now, create a new data frame, df2, that only includes the following variables: btract, BldgClass, LotArea, BldgArea, ComArea, ResArea, OfficeArea, RetailArea, GarageArea, StrgeArea, FactryArea, OtherArea, AreaSource, NumBldgs.

```python
df2 = df1[['btract', 'bldgclass', 'lotarea', 'bldgarea', 'comarea', 'resarea', 'officearea',
         'retailarea', 'garagearea', 'strgearea', 'factryarea', 'otherarea', 'areasource', 'numbldgs']]
df2.head()
```

### 4.	Finally, in another new data frame, df3 use a groupby.sum function using the btract variable in df2 to sum the total type of each area in each census tract.

```python
df3 = df2.groupby('btract')['lotarea','bldgarea','comarea','resarea','officearea',
                            'retailarea','garagearea','strgearea','factryarea','otherarea'].sum()
df3.head()
```

### 5.	This table contains the total type of each area for all census tracts in the city. Export your results to a CSV. Indicate the year in the file name.

```python
df3.to_csv('2010.csv')
```

### 6.	Now, using df2, create a new data frame, df4, that includes only the rows where the building class variable is equal to E1, E3, E4, or E9.  These are all of the buildings classified as warehouses in the city. 

```python
df4 = df2[df2['bldgclass'].isin(['E1', 'E3', 'E4', 'E9'])]
df4.head()
```

### 7.	Repeat step 4 and 5 for this data frame.  The resulting table will include the total type of each area in each census tract only for buildings classified as warehouses. Indicate the year and WH only in the file name.

```python
df5 = df4.groupby('btract')['lotarea','bldgarea','comarea','resarea','officearea',
                            'retailarea','garagearea','strgearea','factryarea','otherarea'].sum()
df5.head()
```

```python
df5.to_csv('2010_WH.csv')
```
